```python
tot_existing = 0
tot_attired = 0
for index, row in df.iterrows():
    if row["Attrition_Flag"] == "Existing Customer":
        tot_existing += 1
    else:
        tot_attired += 1
print("total existing : ", tot_existing)
print("total attired/left : ", tot_attired)
```

```python
8500/1627
```

```python
store_exist = [0,0,0,0,0,0,0,0]
store_attir = [0,0,0,0,0,0,0,0]
for index, row in df.iterrows():
    if row["Attrition_Flag"] == "Existing Customer":
        store_exist[0] = store_exist[0] + row["Customer_Age"]
        store_exist[1] = store_exist[1] + row["Credit_Limit"]
        store_exist[2] = store_exist[2] + row["Dependent_count"]
        store_exist[3] = store_exist[3] + row["Total_Revolving_Bal"]
        store_exist[4] = store_exist[4] + row["Total_Trans_Amt"]
        store_exist[5] = store_exist[5] + row["Total_Trans_Ct"]
        store_exist[6] = store_exist[6] + row["Total_Ct_Chng_Q4_Q1"]
        store_exist[7] = store_exist[7] + row["Avg_Utilization_Ratio"]
    else:
        store_attir[0] = store_attir[0] + row["Customer_Age"]
        store_attir[1] = store_attir[1] + row["Credit_Limit"]
        store_attir[2] = store_attir[2] + row["Dependent_count"]
        store_attir[3] = store_attir[3] + row["Total_Revolving_Bal"]
        store_attir[4] = store_attir[4] + row["Total_Trans_Amt"]
        store_attir[5] = store_attir[5] + row["Total_Trans_Ct"]
        store_attir[6] = store_attir[6] + row["Total_Ct_Chng_Q4_Q1"]
        store_attir[7] = store_attir[7] + row["Avg_Utilization_Ratio"]
print("Attired Customer", " Existing Customer", " % difference")
st = ["Avg Cust Age", "Avg Credit lim", "Dependent Count", "Total Revolving Bal", "Tot Transac amt", "Tot Trans Ct", "Tot Ct Chng", "Avg Utilization ratio"]
for i in range(0,len(store_exist)):
    if i in [0,1]:
        print(st[i], " ",round(store_attir[i]/tot_attired), " ",round(store_exist[i]/tot_existing), " ", round(abs(round(store_exist[i]/tot_existing)-round(store_attir[i]/tot_attired))/round(store_exist[i]/tot_existing)*100), "%")
    else:
        print(st[i], " ",round(store_attir[i]), " ", round(store_exist[i]), " ", round(abs(round(store_exist[i]) - round(store_attir[i]))/round(store_exist[i])*100), "%")
```

```python
Cat = ["Attrition_Flag", "Gender", "Education_Level", "Marital_Status", "Income_Category", "Card_Category"]
```

```python
# Encode only the categorical columns
encoded_df = pd.get_dummies(df, columns=Cat)

# Find only the new dummy columns (those created from Cat)
new_dummy_cols = [col for col in encoded_df.columns if any(cat in col for cat in Cat)]

# Convert only the dummy columns to integer
encoded_df[new_dummy_cols] = encoded_df[new_dummy_cols].astype(int)
```
